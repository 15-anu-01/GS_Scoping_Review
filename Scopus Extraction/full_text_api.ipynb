```python
# --- CONFIG ---
API_KEY = "04ba27c177017665d0d52733e44f680b"
TOKEN = "6bb7e9d03eba600a0bfbf269c4d9c4b0"
EXCEL_FILE = 'D:/GS/final_scoping_review/final_excels/2024.xlsx'
OUTPUT_DIR = 'D:/GS/final_scoping_review/final_excels/final_full_text_1_2024/DOI'

os.makedirs(OUTPUT_DIR, exist_ok=True)

# Function to get full-text article by DOI
def get_full_text_by_doi(doi):
    url = f"https://api.elsevier.com/content/article/doi/{doi}?view=FULL"
    headers = {
        "X-ELS-APIKey": API_KEY,
        "X-ELS-Insttoken": TOKEN,
        "Accept": "application/xml"
    }
    
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        # Save the full-text as XML
        filename = f"{OUTPUT_DIR}/{doi.replace('/', '_')}.xml"
        with open(filename, "wb") as f:
            f.write(response.content)
        print(f"✅ Full-text for doi {doi} saved.")
    else:
        print(f"❌ Failed to retrieve full text for doi {doi}. Status Code: {response.status_code}")

# Read the Excel file into a DataFrame
df = pd.read_excel(EXCEL_FILE)

# Loop through each DOI and process it
for index, row in df.iterrows():
    doi = row.get('DOI')
    if pd.notna(doi):
        get_full_text_by_doi(doi)
    else:
        print(f"⚠️ doi missing in row {index}. Skipping.")

print("All processing completed.")
```
