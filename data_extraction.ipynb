```python
import boto3
import json
import pandas as pd

bedrock = boto3.client('bedrock-runtime', region_name = 'us-east-1')
```

```python
df = pd.read_excel("D:/GS/final_scoping_review/final_excels/yearly_grouped_excels/2023_2024.xlsx")
output = "D:/GS/final_scoping_review/data_extractions/stage1/2023_2024.xlsx"
```

```python
# data extraction prompt

def extract_data(abstract, result):
    
    # payload
    payload = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 20000,
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": f"""
Human: Using the abstract and the results, extract the following information:  
1. What are the core smart technologies used in the paper?
2. which healthcare problem does the paper focus on?  
3. Which country/state/place the paper focus on?
No additional explanations or reasons for any case.

If any information is not available, answer "None".

Return the output as:  
Technology: ...,  
Problem: ...,
Geographic: ...,

Abstract: {abstract}
Result: {result}
                        """
                    }
                ]
            }
        ]
    }
    
    # Invoke model
    response = bedrock.invoke_model(
            body=json.dumps(payload),
            modelId='anthropic.claude-3-5-sonnet-20240620-v1:0',
            accept='application/json',
            contentType='application/json'
        )
    

    # full_response = json.loads(response['body'].read().decode('utf-8'))

    full_response = json.loads(response['body'].read().decode('utf-8'))
    assistant_message = full_response['content'][0]['text']

    extracted_data = {
        "Technology": "None",
        "Problem": "None",
        "Geographic": "None"
    }

    lines = assistant_message.split("\n")

    for line in lines:
        if line.startswith("Technology:"):
            extracted_data["Technology"] = line.split(": ")[1].strip()
        elif line.startswith("Problem:"):
            extracted_data["Problem"] = line.split(": ")[1].strip()
        elif line.startswith("Geographic:"):
            extracted_data["Geographic"] = line.split(": ")[1].strip()

    return extracted_data

# processing each abstract and extracting data
results = []
for index, row in df.iterrows():
    abstract = row['Abstract']
    result = row['Results']
    # classification = classify_abstract(abstract)
    extracted_data = extract_data(abstract, result)
    results.append(extracted_data)

    print(f"Processed {index+1}/{len(df)}")

results_df = pd.DataFrame(results)
result_df = pd.concat([df.reset_index(drop=True), results_df], axis=1)

# Final save after full run
result_df.to_excel(output, index=False)

print("Extraction completed and saved to '2023_2024.xlsx'")
```
